```python
def dw(data):
    ols_res = OLS(data, np.ones(len(data))).fit() # Fitting a regression model
    durbin_watson_value = durbin_watson(ols_res.resid) # running durbin watson test on the residuals
    durbin_watson_value = round( durbin_watson_value, 2) # rounding
    print(f'Durbin Watson Value of: {durbin_watson_value}')
    df = pd.DataFrame([{'y':0, 'dw': durbin_watson_value}])
    max_val = max(2.5,durbin_watson_value+0.25) # Plotting dimensions
    min_val = min(durbin_watson_value-.25,-.5)
    ax = df.plot(
        kind='scatter', x='dw',y='y', color='red',
        xlim=( min_val , max_val),
        ylabel='',
        yticks=[],
    )
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.stats import boxcox
from scipy.special import inv_boxcox

# Assuming midwest_df is already loaded and indexed by date
# Example: midwest_df = pd.read_csv('path_to_your_file.csv', parse_dates=['date_column'], index_col='date_column')

# Perform Box-Cox transformation
y_transf, lam = boxcox(midwest_df['w_mean sales price ($)'])
midwest_df['transformed_price'] = y_transf

# Verify the Box-Cox transformation
print(midwest_df[['w_mean sales price ($)', 'transformed_price']].head())

# Calculate the split index for 80-20 split
split_index = int(len(midwest_df) * 0.8)

# Split the data into training and test sets
midwest_train = midwest_df.iloc[:split_index]
midwest_test = midwest_df.iloc[split_index:]

# Fit the SARIMAX model on the transformed training data
model = SARIMAX(midwest_train['transformed_price'], order=(1, 1, 1), seasonal_order=(1, 1, 0, 4))
model_fit = model.fit()

# Forecasting
forecast_steps = len(midwest_test)  # Number of steps to forecast (same as length of test data)
forecast = model_fit.get_forecast(steps=forecast_steps)
forecast_mean_transformed = forecast.predicted_mean
forecast_ci_transformed = forecast.conf_int()

# Inverse Box-Cox transformation to get back to original scale
forecast_mean = inv_boxcox(forecast_mean_transformed, lam)
forecast_ci_lower = inv_boxcox(forecast_ci_transformed.iloc[:, 0], lam)
forecast_ci_upper = inv_boxcox(forecast_ci_transformed.iloc[:, 1], lam)

# Calculate the errors
mse = mean_squared_error(midwest_test['w_mean sales price ($)'], forecast_mean)
mae = mean_absolute_error(midwest_test['w_mean sales price ($)'], forecast_mean)
mape = np.mean(np.abs((midwest_test['w_mean sales price ($)'] - forecast_mean) / midwest_test['w_mean sales price ($)'])) * 100

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'MAPE: {mape}%')

# Applying dark theme
plt.style.use('dark_background')

# Plotting the forecast
plt.figure(figsize=(10, 6))
plt.plot(midwest_train.index, midwest_train['w_mean sales price ($)'], label='Observed (Train)', color='lightblue')
plt.scatter(midwest_test.index, midwest_test['w_mean sales price ($)'], label='Observed (Test)', color='orange', marker='o')
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='lightgreen')
plt.fill_between(forecast_mean.index, forecast_ci_lower, forecast_ci_upper, color='pink', alpha=0.3)
plt.xlabel('Date', color='white')
plt.ylabel('Mean Sale Price', color='white')
plt.title('SARIMAX Forecast with Test Data Points', color='white')
plt.legend()
plt.grid(True, color='gray')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sktime.forecasting.arima import AutoARIMA
import matplotlib.pyplot as plt

# Assuming midwest_train and midwest_test are already loaded and indexed by date

# Fit the AutoARIMA model
arima_model = AutoARIMA(D=1, sp=12, d=1)
arima_model.fit(midwest_train)
fh = list(range(1, 1 + len(midwest_test)))
midwest_pred = arima_model.predict(fh=fh)

# Combine all plots into a single figure
plt.figure(figsize=(10, 6))

# Plot training data
plt.plot(midwest_train.index, midwest_train, label='Train', color='blue')

# Plot predictions
plt.plot(midwest_test.index, midwest_pred, label='Predicted', color='red')

# Plot test data
plt.plot(midwest_test.index, midwest_test, label='Test', color='white')

plt.xlabel('Date')
plt.ylabel('Mean Sale Price')
plt.title('ARIMA Forecast with Train, Test, and Predicted Data')
plt.legend()
plt.grid(True)
plt.show()
```
