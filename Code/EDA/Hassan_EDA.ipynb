```python
import matplotlib.pyplot as plt

num_states = len(df['state'].unique())
num_cols = 5
num_rows = (num_states + num_cols - 1) // num_cols  

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows), sharex=True, sharey=True)

for idx, state in enumerate(df['state'].unique()):
    row_idx = idx // num_cols
    col_idx = idx % num_cols
    ax = axes[row_idx, col_idx]

    state_data = df[df['state'] == state]
    ax.plot(state_data.index, state_data['new construction mean sales price ($)'], label=state)
    ax.set_title(state)
    ax.grid(True)

for idx in range(num_states, num_rows * num_cols):
    row_idx = idx // num_cols
    col_idx = idx % num_cols
    axes[row_idx, col_idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the columns you want to include in the histograms
columns_to_plot = [
    'new construction sales count',
    'mean days to pending',
    'median sales price ($)',
    'percent sold above list (%)',
    'zillow home value index',
    'zillow observed rent index'
]

# Create a new DataFrame with only the selected columns
df_subset = df[columns_to_plot]

# Determine the number of rows and columns for subplots
num_cols = 3
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols  # Ensure enough rows to display all histograms

# Create subplots with non-overlapping layout
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes array for easier iteration

# Plot histograms for each selected column
for i, col in enumerate(df_subset.columns):
    ax = axes[i]
    df_subset[col].hist(ax=ax)
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.tick_params(axis='x', rotation=90)  # Rotate x-axis labels

# Hide any unused subplots
for i in range(len(df_subset.columns), num_rows * num_cols):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the columns with "$" sign in their name
price_columns = [col for col in df.columns if '$' in col]

# Create a new DataFrame with only the selected price columns
df_price_subset = df[price_columns]

# Determine the number of rows and columns for subplots
num_cols = 3
num_rows = (len(price_columns) + num_cols - 1) // num_cols  # Ensure enough rows to display all histograms

# Create subplots with non-overlapping layout
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes array for easier iteration

# Plot histograms for each selected price column
for i, col in enumerate(df_price_subset.columns):
    ax = axes[i]
    df_price_subset[col].hist(ax=ax, orientation='horizontal')  # Set orientation to horizontal
    ax.set_title(col)
    ax.set_xlabel('Frequency')
    ax.set_ylabel(col)  # Set the price column as the y-axis label

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axes for subplots
num_cols = 3
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 6 * num_rows))
axes = axes.flatten()

# Iterate over each column in columns_to_plot
for i, col in enumerate(columns_to_plot):
    # Melt the DataFrame to long format to have 'month-year' as the x-axis variable
    df_melted = df.melt(id_vars=['month-year', 'state'], value_vars=[col], var_name='metric', value_name='value')

    # Plot time series for the current column with states as different colors
    sns.lineplot(data=df_melted, x='month-year', y='value', hue='state', style='state', markers=True, dashes=False, ax=axes[i])
    axes[i].set_xlabel('Month-Year')
    axes[i].set_ylabel('Value')
    axes[i].set_title(f'Time Series for {col} with States as Different Colors')
    axes[i].legend().remove()  # Remove individual legend for each subplot
    axes[i].tick_params(axis='x', rotation=45)

# Add a single legend below the subplots
fig.legend(title='State', bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=5)

# Adjust layout to increase the area of the entire figure
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
